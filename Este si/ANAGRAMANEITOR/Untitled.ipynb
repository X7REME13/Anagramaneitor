```python
archivo = 'dicc.txt'
```

```python
with open(archivo, encoding='utf8') as f:
    lines = f.readlines()
```

```python
lines
```

```python
lines = [char.replace('\n','') for char in lines]
```

```python
def es_anagrama_facil(cad1, cad2):
    if cad1 == cad2 or len(cad1) != len(cad2): return False
    list1 = list(cad1)
    list2 = list(cad2)
    cont = 0
    for i, letra2 in enumerate(list1):
        for j, letra1 in enumerate(list2):
            if letra1 == letra2:
                list2[j] = '*'
                cont += 1
                break

    return len(cad1) == cont
            
```

```python
es_anagrama_facil('amor', 'roma')
```

```python
def criterio(e):
    return len(e)


lines.sort(key=criterio)
```

```python
lines
```

```python
indexador = {}
anterior = 0
for i, line in enumerate(lines):
    if len(line) > anterior:
        indexador[f"{len(line)}"] = i
    
```

```python
indexador['0']=0
```

```python
indexador
```

```python
dicc_anagrama = []
encontre = 0
largo = len(lines)
for i,palabra in enumerate(lines):
    print(f"Porcentaje: {i/largo * 100} | encuentros: {encontre}/{i}", end='\r')
    for palabra_comp in lines[indexador[f"{len(palabra)-1}"]:indexador[f"{len(palabra)}"]]:
        #print(palabra, palabra_comp)
        if es_anagrama_facil(palabra, palabra_comp):
            dicc_anagrama.append(palabra)
            encontre+=1
            break
        
```

```python
dicc_anagrama
```

```python
 with open("diccionario_anagrama_JUPYTER.txt","w", encoding='utf-8') as filesave:
    for line in dicc_anagrama:
        filesave.write(line + '\n')
```
